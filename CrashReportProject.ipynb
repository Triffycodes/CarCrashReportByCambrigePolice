```python
#using fuzzywuzzy we are trying to remove the values which are matching with each other with a percentage of 65%,by this way we can remove the
#duplicate street names
import fuzzywuzzy 
from fuzzywuzzy import process

def replace_matches_in_column(df, column, string_to_match, min_ratio = 65):
    # get a list of unique strings
    strings = df[column].unique()
    
    # get the top 5 closest matches to our input string
    matches = fuzzywuzzy.process.extract(string_to_match, strings, 
                                         limit=5, scorer=fuzzywuzzy.fuzz.token_sort_ratio)
# only get matches with a ratio > 65
    close_matches = [matches[0] for matches in matches if matches[1] >= min_ratio]
# get the rows of all the close matches in our dataframe
    rows_with_matches = df[column].isin(close_matches)
    
     # replace all rows with close matches with the input matches 
    df.loc[rows_with_matches, column] = string_to_match
    
    # let us know the function's done

street=df['street_name'].astype('str')
for i in range(0,street.size):
    replace_matches_in_column(df=df, column='street_name', string_to_match=street[i])
 
```

##### Mutual Information (Information gain)

```python
#calculating impurity for information gain
def impurity_cal(feature,criteria):

    probs = feature.value_counts(normalize=True)
    
    if criteria == 'entropy':
        impurity = -1 * np.sum(np.log2(probs) * probs)
    elif criteria == 'gini':
        impurity = 1 - np.sum(np.square(probs))
    else:
        raise ValueError('Invalid criteria')
        
    return(round(impurity, 3))
```

```python
#Calculating Information gain
def comp_infogain(df, target, descriptive_feature, split_criterion):
    print('target feature:', target)
    print('descriptive_feature:', descriptive_feature)
    print('split criterion:', split_criterion)
            
    target_entropy = impurity_cal(df[target], split_criterion)

    # entropy_list to store  entropy of each partition
    # weight_list to store the relative number of observations of each partition
    #gini_list to store gini_index of each partition
    entropy_list = list()
    weight_list = list()
    
    # loop over each level of the descriptive feature
    # to partition the dataset with respect to that level
    # and compute the entropy and the weight of the level's partition
    for level in df[descriptive_feature].unique():
         df_feature_level = df[df[descriptive_feature] == level]
         entropy_level = impurity_cal(df_feature_level[target], split_criterion)
         entropy_list.append(round(entropy_level, 3))
         weight_level = len(df_feature_level) / len(df)
         weight_list.append(round(weight_level, 3))
    print('impurity of partitions:', entropy_list)
    print('weights of partitions:', weight_list)
    feature_impurity = np.sum(np.array(entropy_list) * np.array(weight_list))
    print(' impurity:', feature_impurity)
    information_gain = target_entropy - feature_impurity
    print('information gain:', information_gain)
    print('====================')
    return(information_gain)    
```

```python
split_criterion = 'entropy'
gainn={}
for feature in df.drop(columns='Accident_predict').columns:
    feature_info_gain = comp_infogain(df, 'Accident_predict', feature, split_criterion)
    gainn[feature]=feature_info_gain
```

##### GINI INDEX

```python
def gini_index(df, target, descriptive_feature, split_criterion):
    print('target feature:', target)
    print('descriptive_feature:', descriptive_feature)
    print('split criterion:', split_criterion)
            
    target_entropy = impurity_cal(df[target], split_criterion)

    # entropy_list to store  entropy of each partition
    # weight_list to store the relative number of observations of each partition
    #gini_list to store gini_index of each partition
    entropy_list = list()
    weight_list = list()
    gini_list=list()
    
    # loop over each level of the descriptive feature
    # to partition the dataset with respect to that level
    # and compute the entropy and the weight of the level's partition
    for level in df[descriptive_feature].unique():
       df_feature_level = df[df[descriptive_feature] == level]
       gini_level = impurity_cal(df_feature_level[target], split_criterion)
       gini_list.append(round(gini_level, 3))
       weight_level = len(df_feature_level) / len(df)
       weight_list.append(round(weight_level, 3))
    print('impurity of partitions:', gini_list)
    print('weights of partitions:', weight_list)
    feature_impurity = np.sum(np.array(gini_list) * np.array(weight_list))
    print('impurity:', feature_impurity)
    gini_index = target_entropy - feature_impurity
    print('Gini Index:', gini_index)
    print('====================')
    return(gini_index)    
```
